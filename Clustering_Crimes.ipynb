## Hierarchical Clustering

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import sklearn as skl 
import seaborn as sn
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
crime = pd.read_csv("C:\\Users\\User\\Documents\\Datascience_Assignments\\Assignment 7 (Clustering)\\crime_data.csv")
crime
```

### Normalization Function

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return x
```

```python
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

### Dendrogram

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm, method = 'complete'))
```

```python
hc1 = AgglomerativeClustering(n_clusters= 4, affinity='euclidean',linkage = 'complete')
```

```python
y_hc1 = hc1.fit_predict(df_norm)
y_hc1
```

```python
clusters = pd.DataFrame(y_hc1, columns = ['Clusters'])
clusters
```

```python
crime['h_clusterid'] = hc1.labels_
crime
```

## K-means clustering

```python
crime = pd.read_csv("C:\\Users\\User\\Documents\\Datascience_Assignments\\Assignment 7 (Clustering)\\crime_data.csv")
crime
```

```python
df_norm = norm_func(crime.iloc[:,1:])
```

```python
k = list(range(2,15))
k
TWSS = []  
for i in k:
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_norm)
    WSS = []  
    for j in range(i):
        WSS.append(sum(cdist(df_norm.iloc[kmeans.labels_==j,:],kmeans.cluster_centers_[j].reshape(1,df_norm.shape[1]),"euclidean")))
    TWSS.append(sum(WSS))
```

### Scree plot 

```python
plt.plot(k,TWSS, 'ro-');plt.xlabel("No_of_Clusters");plt.ylabel("total_within_SS");plt.xticks(k)
```

```python
model = KMeans(n_clusters = 4).fit(df_norm)
model.cluster_centers_
```

```python
model.labels_ 
md=pd.Series(model.labels_)   
crime['clust'] = md  
crime
```

```python
crime.value_counts('clust')
```

```python
crime.iloc[:,1:5].groupby(crime.clust).mean()
```

## DBscan clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime = pd.read_csv("C:\\Users\\User\\Documents\\Datascience_Assignments\\Assignment 7 (Clustering)\\crime_data.csv")
crime
```

```python
crime.drop(['Unnamed: 0'], axis = 1, inplace = True) 
crime
```

```python
array = crime.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
dbscan = DBSCAN(eps = 1, min_samples = 5)
dbscan.fit(X)
```

```python
dbscan.labels_
```

```python
cl = pd.DataFrame(dbscan.labels_,columns = ['cluster'])
cl
```

```python
ds = pd.concat([crime,cl],axis = 1)
ds
```

```python
ds.value_counts('cluster')
```

```python
ds.iloc[:,0:4].groupby(ds.cluster).mean()
```

#### Plot Clusters

```python
plt.figure(figsize=(10, 7))  
plt.scatter(ds['cluster'],ds['UrbanPop'], c = dbscan.labels_) 
```
